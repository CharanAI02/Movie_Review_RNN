```python
from tensorflow.keras.preprocessing.text import one_hot
```

```python
### sentences
sent=[  'the glass of milk',
     'the glass of juice',
     'the cup of tea',
    'I am a good boy',
     'I am a good developer',
     'understand the meaning of words',
     'your videos are good',]
```

```python
sent
```

```python
#Define the vocabulary size

voc_size=10000
```

```python
## One Hot Representation for every word

one_hot_rep=[one_hot(words,voc_size)for words in sent]
one_hot_rep
```

```python
#WOrd Embedding Representation

from tensorflow.keras.layers import Embedding
from tensorflow.keras.models import Sequential
from tensorflow.keras.utils import pad_sequences   #To make all sent to equal size
import numpy as np
```

```python
sent_length=8
embedded_docs=pad_sequences(one_hot_rep,padding='pre',maxlen=sent_length)
print(embedded_docs)
```

```python
#Feature representation
dim=10
```

```python
model = Sequential()
model.add(Embedding(input_dim=voc_size,   # Vocabulary size
                    output_dim=dim,     # Embedding dimension
                    input_length=sent_length))  # Length of input sequences

# Compile the model
model.compile(optimizer='adam', loss='mse')
```

```python
model.summary()
```

```python
model.predict(embedded_docs)
```
